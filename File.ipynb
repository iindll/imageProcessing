```python
#Threshold
import os,cv2
image = cv2.imread("./data/horse.jpeg")
gray_img = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
ret ,Threshold = cv2.threshold(gray_img,80,255,cv2.THRESH_BINARY)
Threshold = cv2.blur(Threshold , (10,10))
ret ,Threshold = cv2.threshold(gray_img,180,255,cv2.THRESH_BINARY)

cv2.imshow("image",Threshold)

cv2.waitKey(0)

```

```python
import cv2
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import time

webcam = cv2.VideoCapture(0)

try:
    while True:   # infinite loop
        ret, frame = webcam.read()
    
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        clear_output(wait=True)  
        plt.imshow(frame_rgb)
        plt.axis("off")
        display(plt.gcf())

        time.sleep(0)   

except KeyboardInterrupt:
    print("‚èπ Stopped manually")

webcam.release()
```
